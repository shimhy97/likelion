#### 모델 작성

````
# 딥러닝을 구동하는 데 필요한 케라스 함수 및 필요한 라이브러리 불러오기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
````

````
# 실행할 때마다 같은 결과를 출력하기 위해 설정


seed = 3
np.random.seed(seed)
tf.random.set_seed(seed)
````

````
# 수술 환자 데이터를 가져옴

data_set= np.loadtxt(drive_path + "ThoraricSurgery.csv", delimiter=",")      #넘파이에서 텍스트 파일 가져오는 함수. delimeter 속성으로 구분자 설정 가능.

type(data_set)
````

````
# 환자의 기록과 수술 결과를 X, y로 구분하여 저장
# 신경망에 학습시킬 때 사용하는 특성은 총 몇개일까?

X = data_set[:,:17]
y=data_set[:,17]
````

````
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

# 여기까진 똑같네! 데이터 불러오고, 특성, 타깃 분리하고 학습용 테스트용 나누기!

````

````
# 딥러닝 구조를 결정하는 부분.
# 실제 프로젝트에선 이 부분이 상당히 복잡해진다!!!!
# 모델을을 붙이거나, 층을 여러개 만들거나.. 어떻게 모델을 구상할건데???

model = Sequential()

# 30은 뉴런의 갯수이다. 층은 두개를 만들었다
# input_dim = (하나의 벡터를 구성하는) 차원!! 매우 중요함 // activation 은 활성함수? 를 적용하는 곳이다.
# 그냥 처음에 넣는 특성갯수가 17개 아닌가? 모르겠다
# 'relu'는 뭐나면 .. 시그모이드 함수 기억남? 거기서 0이하로 안내려가게 하는 함수래, 한마디로 활성함수!!

model.add(Dense(30, input_dim=17, activation='relu'))      
model.add(Dense(1,activation='sigmoid'))

#결국 조립만 잘 하면 된다.
#모델의 구조를 만들어 보았다.
````

층(Layer) : 핵심 구성요소로 일종의 데이터 처리 필터 (어떤 데이터가 들어가면 더 유용한 형태로 출력)
* 주어진 문제에 더 의미 있는 표현을 입력된 데이터로부터 추출

````
# 학습 실행
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])       # 로스, 옵티마이저 설정. 
model.fit(X_train, y_train, epochs=100)                             #epochs = 반복횟수 설정

````

> compile() : 학습 프로세스를 설정하는 기능
 
> fit() : 학습 데이터로 학습하는 기능
 
complie(손실함수 종류, 최적화(오차감소) 방법, 평가 방법)
 
* fit(입력 데이터, 입력 라벨, 학습 반복 횟수(30개 층에 대해 한번씩 계산), 한번에 처리할 데이 터 수)

* 손실함수(loss function) : 오차 계산 방법 (MSE (평균제곱오차), CEE (교차 엔트로피 오차) 등)

* 최적화 방법(Optimizer) : 정답을 찾아가는 방법 (SGD (확률적 경사하강법), 모멘텀, AdaGrad, Adam 등)

* 훈련과 테스트 과정을 모니터링할 지표 : 여기에서는 정확도만 고려

````
# 결과로 두개가 나오는데, 0번 인덱스가 가중치값, 1번 인덱스가 정확도(Accuracy)이다.

print("Accuracy: %.4f" % model.evaluate(X_test, y_test)[1] )
````

## 구성요소

### 텐서(tensor)

* 신경망을 위한 데이터 표현

* 텐서(tensor)를 기본 구조로 사용

* 텐서(tensor) : 데이터를 위한 컨테이너 (숫자를 위한 컨테이너)
 - 임의의 차원 개수를 가지는 행렬의 일반화된 모습 (다차원 배열)
 - 차원(dimension)을 축(axis)이라고 부름


텐서의 실제 사례
* (samples, features) : 벡터 배열 데이터
* (samples, timesteps, features) : 시계열 데이터 또는 시퀀스 데이터
* (samples, height, width, channels) : 이미지 데이터
* (samples, frames, height, width, channels) : 동영상 데이터


핵심 속성 3가지
* 축의 개수(rank)
* 크기(shape) : 텐서의 각 축을 따라 얼마나 많은 차원이 있는지를 나타내는 파이썬의 튜플
* 데이터 타입

* 하나의 숫자 값
* 넘파이의 float32, float64 타입의 숫자 배열
* 랭크(rank) : 텐서의 축 개수

````
def element_add(x, y):
  assert len(x.shape) == 2  # x, y는 2D 배열(행렬)
  assert x.shape == y.shape

  x = x.copy()  # 입력 배열 원본을 변경시키지 않기 위해 복사
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] += y[i, j]
  
  return x
````

````
def element_relu(x):
  assert len(x.shape) == 2  # x는 2D 배열(행렬)

  x = x.copy()  # 입력 배열 원본을 변경시키지 않기 위해 복사
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] = max(x[i, j], 0)   # 0이하는 0으로 저장
  
  return x
````

````
#차원이 다를때, 곱셈 연산이 어떻게 일어나는지 알려준다.
#차원이 작은쪽이 차원이 큰 쪽에 맞춰서, 반복되는구만

def element_relu_matrix_and_vector(x, y):
  assert len(x.shape) == 2  # x는 2D 배열(행렬)
  assert len(y.shape) == 1  # x는 벡터
  assert x.shape[1] == y.shape[0]

  x = x.copy()  # 입력 배열 원본을 변경시키지 않기 위해 복사
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] += y[j]
  
  return x
````

````
def vector_dot(x, y):
  assert len(x.shape) == 1  # x는 벡터
  assert len(y.shape) == 1  # y는 벡터
  assert x.shape[0] == y.shape[0]

  z = 0.
  for i in range(x.shape[0]):
    z += x[i] * y[i]
  
  return z
````

````
def matrix_vector_dot(x, y):
  assert len(x.shape) == 2  # x는 행렬
  assert len(y.shape) == 1  # y는 벡터
  assert x.shape[1] == y.shape[0]    #행렬곱이 정의되도록 사이즈 맞춤

  z = np.zeros(x.shape[0]) # x행과 같은 크기의 0으로 채워진 벡터
  for i in range(x.shape[0]):
    z[i] = vector_dot(x[i, :], y)
  
  return z
````

````
def matrix_dot(x, y):
  assert len(x.shape) == 2  # x는 2차원행렬
  assert len(y.shape) == 2  # y는 2차원행렬
  assert x.shape[1] == y.shape[0]     #행렬곱이 정의되도록 사이즈 맞춤

  z = np.zeros( (x.shape[0], y.shape[1]) ) # 0으로 채워진 특정 크기의 벡터
  for i in range(x.shape[0]):
    for j in range(y.shape[1]):
      row = x[i, :]
      column = y[:, j]
      z[i, j] = vector_dot(row, column)
  
  return z
````

##전체코드

````
# 딥러닝을 구동하는 데 필요한 케라스 함수를 가져오기
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

x1 = np.random.uniform(0.0, 1.0, size=100).reshape(-1,1)
x2 = np.random.uniform(0.0, 1.0, size=100).reshape(-1,1)
X = np.concatenate( (x1, x2), axis=1)

y = (12 * x1) + (-5 * x2) + 3       # y = W1 * x1 + w2 * x2 + b // 참값

# 딥러닝 구조를 결정 (모델을 설정하고 실행하는 부분)
# 선형모델 1개로 구성된 신경망을 생성
model = Sequential()
model.add(Dense(1, input_dim=2, name = 'linear_dense'))     # dense에 이름 설정도 가능
model.compile(loss="mse", optimizer="SGD")

model.fit(X, y, epochs=100)         # 학습실행

model.summary()                      # 요약정보
model.get_layer('linear_dense').get_weights()  # 가중치 확인
````
